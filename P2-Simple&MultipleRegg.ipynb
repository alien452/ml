**1).THIS IS OF DATASET FROM SKLEARN**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load California Housing dataset
housing = fetch_california_housing()
X = housing.data     # multiple features
y = housing.target   # target variable (median house value)

# For simplicity, select only two features to visualize easily
# Let's pick 'MedInc' (median income) and 'AveRooms' (average rooms)
X_selected = X[:, [0, 3]]

# Split data into train/test
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Train the Multiple Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Print coefficients and evaluation metrics
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared Score:", r2_score(y_test, y_pred))

# Visualize Actual vs Predicted house values
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred, color='green', alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')
plt.xlabel('Actual Median House Value')
plt.ylabel('Predicted Median House Value')
plt.title('Actual vs Predicted Median House Values')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Sample dataset
data = {
    'Hours_Studied': [10, 9, 2, 15, 10, 16, 11, 16, 5, 8],
    'Score': [85, 80, 40, 95, 60, 98, 85, 100, 45, 70]
}
df = pd.DataFrame(data)

# Feature and target
X_simple = df[['Hours_Studied']]  # Single feature
y = df['Score']

# Train/test split
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y, test_size=0.3, random_state=42)

# Model training
model_simple = LinearRegression()
model_simple.fit(X_train_s, y_train_s)

# Predict
y_pred_s = model_simple.predict(X_test_s)

# Evaluation
print("Simple Linear Regression:")
print("Coefficient:", model_simple.coef_)
print("Intercept:", model_simple.intercept_)
print("MSE:", mean_squared_error(y_test_s, y_pred_s))
print("R2 Score:", r2_score(y_test_s, y_pred_s))
print()

# Print actual vs predicted values side by side
print("Actual vs Predicted values:")
for actual, pred in zip(y_test_s, y_pred_s):
    print(f"Actual: {actual:.2f}, Predicted: {pred:.2f}")

# Plot Simple Linear Regression results
plt.figure(figsize=(8,5))
plt.scatter(X_test_s, y_test_s, color='blue', label='Actual')
plt.plot(X_test_s, y_pred_s, color='red', linewidth=2, label='Predicted')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Sample dataset
data = {
    'Hours_Studied': [10, 9, 2, 15, 10, 16, 11, 16, 5, 8],
    'Attendance': [90, 80, 60, 95, 70, 98, 85, 100, 50, 75],
    'Score': [85, 80, 40, 95, 60, 98, 85, 100, 45, 70]
}
df = pd.DataFrame(data)

# Features and target
X_multiple = df[['Hours_Studied', 'Attendance']]  # Two features
y = df['Score']

# Train/test split
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multiple, y, test_size=0.3, random_state=42)

# Model training
model_multiple = LinearRegression()
model_multiple.fit(X_train_m, y_train_m)

# Predict
y_pred_m = model_multiple.predict(X_test_m)

# Evaluation
print("Multiple Linear Regression:")
print("Coefficients:", model_multiple.coef_)
print("Intercept:", model_multiple.intercept_)
print("MSE:", mean_squared_error(y_test_m, y_pred_m))
print("R2 Score:", r2_score(y_test_m, y_pred_m))
print()

# Print actual vs predicted values using zip
print("Actual vs Predicted values:")
for actual, pred in zip(y_test_m, y_pred_m):
    print(f"Actual: {actual:.2f}, Predicted: {pred:.2f}")

# Plot Actual vs Predicted Scores for Multiple Regression
plt.figure(figsize=(8,5))
plt.scatter(y_test_m, y_pred_m, color='green', edgecolor='k', alpha=0.7)
plt.plot([min(y_test_m), max(y_test_m)], [min(y_test_m), max(y_test_m)], 'r--')  # 45-degree reference line
plt.xlabel('Actual Score')
plt.ylabel('Predicted Score')
plt.title('Multiple Linear Regression\nActual vs Predicted Scores')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# 1) Retrieve the Dataset (defined manually here)
data = {
    'Hours_Studied': [10, 9, 2, 15, 10, 16, 11, 16, 5, 8],
    'Attendance': [90, 80, 60, 95, 70, 98, 85, 100, 50, 75],
    'Score': [85, 80, 40, 95, 60, 98, 85, 100, 45, 70]
}
df = pd.DataFrame(data)
print("Dataset:")
print(df)

# 2) Prepare features and target
X = df[['Hours_Studied', 'Attendance']]
y = df['Score']

# Split dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the multiple linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the scores for test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print coefficients and metrics
print("\nMultiple Linear Regression Model:")
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")

# Show Actual vs Predicted values
print("\nActual vs Predicted Scores:")
for actual, pred in zip(y_test, y_pred):
    print(f"Actual: {actual:.2f}, Predicted: {pred:.2f}")

# Conclusion
print("\nConclusion:")
print("The model uses Hours Studied and Attendance to predict the Score.")
print("Coefficients indicate the influence of each feature on the score.")
print("The R-squared score near 1 suggests a good fit on this small dataset.")
print("Lower Mean Squared Error means the predictions are close to actual values.")
```
